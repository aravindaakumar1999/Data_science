```python
import os
os.chdir('/home/aravinda/Desktop/MCA/data_modeling')
novel=open('pdfcoffee.com_crooked-house-by-agatha-christie-pdf-free.txt','r').read()
```

```python
words=word_tokenize(novel)
print(words)
```

```python
print(f"The total number of words {len(words)}")
```

```python
print(novel[1000:2000])
```

```python
fdist=FreqDist(words)
fdist.most_common(15)
```

```python
word_no_punc=[]
for word in words:
    if word.isalpha():
        word_no_punc.append(word.lower())
print(f"{len(word_no_punc)}")
```

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
stopwords_list=stopwords.words("english")
print(stopwords_list)
```

```python
clean_words=[]
for word in word_no_punc:
    if word not in  stopwords_list:
        clean_words.append(word)
print(f"{len(clean_words)}")
print(clean_words)
```

```python
fdist=FreqDist(clean_words)
fdist.most_common(15)
```

```python
clean_words_string=" ".join(clean_words)
word_cloud=WordCloud(background_color="green").generate(clean_words_string)
plt.figure(figsize=(12,12))
plt.imshow(word_cloud)
```
